```python
# Problem 4

import time

THICKNESS = 0.00008
fold = THICKNESS
    

def compareTime(time1, time2):
    if (time1==time2):
        print ("Time 1 equals time 2 ")
    elif (time1>time2):
        print ("Time 1  is greater than time 2 ")
    else:
        print ("Time 2 is greater than time 1 ")

# Recording the time of problem 1 that'll be saved in time1 variable
start1 = time.time()
folded_thickness = THICKNESS * 2**43
print ("The thickness of paper folded 43 times is {} meters".format(folded_thickness))
final_time1 = time.time()
time1 = final_time1-start1
print("time 1 : {}[s]".format(time1))

# Recording the time of problem 3 that'll be saved in time2 variable
start2 = time.time()
for i in range(-1, 43):
    fold = fold*2
print ("The thickness of paper folded 43 times is {} meters".format(folded_thickness))
final_time2 = time.time()
time2 = final_time2-start2
print("time 2 : {}[s]".format(time2))

compareTime(time1, time2)

```

```python
# Problem 4
# Increase number of repetitions

import time

THICKNESS = 0.00008
fold = THICKNESS
    

def compareTime(time1, time2):
    if (time1==time2):
        print ("Time 1 equals time 2 ")
    elif (time1>time2):
        print ("Time 1  is greater than time 2 ")
    else:
        print ("Time 2 is greater than time 1 ")

# Recording the time of problem 1 that'll be saved in time1 variable
start1 = time.time()
folded_thickness = THICKNESS * 2**1000
print ("The thickness of paper folded 1000 times is {} meters".format(folded_thickness))
final_time1 = time.time()
time1 = final_time1-start1
print("time 1 : {}[s]".format(time1))

# Recording the time of problem 3 that'll be saved in time2 variable
start2 = time.time()
for i in range(-1, 1000):
    fold = fold*2
print ("The thickness of paper folded 10000 times is {} meters".format(folded_thickness))
final_time2 = time.time()
time2 = final_time2-start2
print("time 2 : {}[s]".format(time2))

compareTime(time1, time2)

```

```python
#Problem 5


THICKNESS = 0.00008
fold = THICKNESS
list_of_folds = []
    
for i in range(-1, 43):
    
    fold = fold*2
    list_of_folds.append(fold)

print(len(list_of_folds))

```

```python
#Problem 6
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [m]")
plt.plot(list_of_folds)
plt.show()
```
